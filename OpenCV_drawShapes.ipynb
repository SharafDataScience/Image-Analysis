# **Title**: OpenCV Image Processing Notebook

# **Description**: This notebook demonstrates the use of OpenCV and matplotlib to create and manipulate images. Key functionalities include creating blank images, drawing shapes, and visualizing images using OpenCV functions. Each code block is thoroughly commented to provide a clear understanding of the processing steps.



```python

import cv2
import numpy as np
from matplotlib import pyplot as plt

```

```python
# **Create a blank image**
# A blank image is created with dimensions 512x512 and a color depth of 3 channels (RGB).
blank_img = np.zeros((512, 512, 3), dtype=np.uint8)  # Blank black image

```

```python
# **Draw a rectangle**
# Drawing a rectangle on the blank image. The rectangle is defined by its top-left and bottom-right corners.
cv2.rectangle(blank_img, (100, 100), (300, 300), (255, 0, 0), thickness=3)  # Blue rectangle with thickness 3


```

```python
# **Draw a circle**
# A circle is drawn with a specified center and radius.
cv2.circle(blank_img, (400, 400), 50, (0, 255, 0), thickness=-1)  # Filled green circle

# **Display the image**
# The image is displayed using matplotlib to visualize the drawn shapes.
plt.imshow(blank_img)  # Convert BGR (OpenCV default) to RGB for matplotlib
plt.title("Rectangle and Circle")
plt.show()

```

```python
# **Define a polygon**
# A polygon is defined using an array of vertices and reshaped to a specific format.
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)  # Polygon vertices
print("Vertices Shape:", vertices.shape)  # Printing the shape of the vertices array

# **Prepare the polygon for drawing**
# The vertices are reshaped into a format compatible with OpenCV's polylines function.
pts = vertices.reshape((-1, 1, 2))
print("Points Shape:", pts.shape)  # Printing the shape after reshaping

# **Draw the polygon**
# The polygon is drawn with specified color and thickness.
cv2.polylines(blank_img, [pts], isClosed=True, color=(255, 0, 255), thickness=5)  # Pink polygon with thickness 5

# **Display the final image**
# The final image with the rectangle, circle, and polygon is displayed.
plt.imshow(blank_img)  # Convert BGR (OpenCV default) to RGB for matplotlib
plt.title("Rectangle, Circle, and Polygon")
plt.show()
```

```python
# add line to plot
cv2.line(blank_img,(100,250),(300,400),(255,255,255),thickness=3)
plt.imshow(blank_img)
```

```python
# add text to plot.
font = cv2.FONT_HERSHEY_TRIPLEX
cv2.putText(img=blank_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
