```python
# Import libraries.
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# open image
img = cv2.imread('/00-puppy.jpg')
print(type(img))
print(img.shape)
plt.imshow(img)
```

```python
# matplotlib RGB , CV2 BGR, then convert to RGB
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(fix_img.shape)
plt.imshow(fix_img)
```

```python
# grey scale image
img_grey = cv2.imread('/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
print(img_grey.shape)
plt.imshow(img_grey, cmap='gray')
```

```python
# resize pic
new_img = cv2.resize(fix_img, (1000, 400))
print(new_img.shape)
plt.imshow(new_img)
```

```python
# resize pic by rate.
new_img = cv2.resize(fix_img, (0,0),fix_img, 0.8,0.2)
print(new_img.shape)
plt.imshow(new_img)
#
```

```python
# Flip the image in three ways
img_flip_horizontally = cv2.flip(fix_img, 1)  # Horizontal flip
img_flip_vertically = cv2.flip(fix_img, 0)    # Vertical flip
img_flip_both = cv2.flip(fix_img, -1)         # Flip both axes

# Create subplots for displaying the images
plt.figure(figsize=(15, 5))

# Original Image
plt.subplot(1, 4, 1)
plt.imshow(fix_img)
plt.title("Original")
plt.axis('off')

# Horizontal Flip
plt.subplot(1, 4, 2)
plt.imshow(img_flip_horizontally)
plt.title("Horizontal Flip")
plt.axis('off')

# Vertical Flip
plt.subplot(1, 4, 3)
plt.imshow(img_flip_vertically)
plt.title("Vertical Flip")
plt.axis('off')

# Both Axes Flip
plt.subplot(1, 4, 4)
plt.imshow(img_flip_both)
plt.title("Both Axes Flip")
plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```
